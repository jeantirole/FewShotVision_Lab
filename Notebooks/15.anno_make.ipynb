```python
import os
import numpy as np
from PIL import Image, ImageDraw
import json
import matplotlib.pyplot as plt
from glob import glob

# Parameters
target_img_id = 7
mask_on = True
bbox_on = False
create_mask = True
mask_save_folder = "/home/eric/srcs/FewShotSeg_Lab/FewShotVision_Lab/SmallObjectFSS/SOFS/Open_Domain_Data/PCB_test/support/mask"

# -- Load Annotations
with open(anno_path, 'r') as f:
    data = json.load(f)

# Get file name for the target image
file_name = [i['file_name'] for i in data['images'] if i['id'] == target_img_id][0]
print("File name:", file_name)

# -- Load Image
for name in glob(os.path.join(folder, "*.jpg")):
    if name.split("/")[-1].split(".")[0] == file_name.split(".")[0]:
        print("Load image:", name)
        img = Image.open(name)

# Get image size
width, height = img.size
print("Image size:", width, height)

# Create an empty binary mask
binary_mask = np.zeros((height, width), dtype=np.uint8)

# -- Process Annotations
for anno in data['annotations']:
    if anno['image_id'] == target_img_id:
        print("Defect category:", anno["category_id"])

        if mask_on and 'segmentation' in anno and anno['segmentation']:
            # Create a segmentation mask
            segmentation_instance = anno['segmentation'][0]
            x = segmentation_instance[0::2]  # x-coordinates
            y = segmentation_instance[1::2]  # y-coordinates
            polygon = list(zip(x, y))

            # Draw the segmentation on the mask
            mask_img = Image.new('L', (width, height), 0)  # Create a blank grayscale image
            ImageDraw.Draw(mask_img).polygon(polygon, outline=1, fill=1)  # Draw the polygon
            binary_mask += np.array(mask_img)  # Add the mask to the binary mask

# Ensure the mask is binary (0 or 1)
binary_mask = (binary_mask > 0).astype(np.uint8)

# -- Save Binary Mask
if create_mask:
    #os.makedirs(folder, exist_ok=True)  # Ensure the folder exists
    mask_file_name = os.path.join(mask_save_folder, f"{file_name.split('.')[0]}_mask.png")
    mask_image = Image.fromarray(binary_mask * 255)  # Convert binary mask to 255 scale for saving
    mask_image.save(mask_file_name)
    print("Binary mask saved as:", mask_file_name)

# -- Optional: Visualize the Binary Mask
plt.figure(figsize=(10, 8))
plt.imshow(binary_mask, cmap='gray')
plt.title("Binary Mask")
plt.axis('off')
plt.show()
```

```python
folder = "/home/eric/data/anomaly_detection_data/files/PCB_1/val"
anno_path = glob(os.path.join(folder, "*.json"))[0]
```

```python
target_img_id = 3 
mask_on = False
bbox_on = True

#-- anno
with open(anno_path, 'r') as f:
    data = json.load(f)
file_name = [ i['file_name'] for i in data['images'] if i ['id'] == target_img_id][0]

display("defect category table :", data['categories'])

#-- img 
for name in glob(os.path.join(folder, "*.jpg")):
    # if int(name.split("/")[-1].split(".")[0]) == target_img_id:
    if name.split("/")[-1].split(".")[0] == file_name.split(".")[0]:
        print("load image : " , name)
        img = PIL.Image.open(name)

plt.figure(figsize=(16,8))
plt.imshow(img)

#-- figure 
plt.figure(figsize=(16,8))
#-- segmentation
for anno in data['annotations']:
    if anno['image_id']==target_img_id:
        #-- category
        print("defect category :", anno["category_id"])
        print("category type : ",[ i for i in data['categories'] if i ['id'] == anno["category_id"]] )
        
        if mask_on: 
            #--- seg mask 
            segmentation_instance = anno['segmentation'][0]
            x = segmentation_instance[0::2]  # Extract every other value starting from the first (x-coordinates)
            y = segmentation_instance[1::2]  # Extract every other value starting from the second (y-coordinates)
            # Add the first point to the end to close the polygon
            x.append(x[0])
            y.append(y[0])
            # vis
            if mask_on == True:
                plt.plot(x, y, marker='o', linestyle='-', color='b')  # Plot lines and points
                plt.fill(x, y, color='skyblue', alpha=0.4)  # Fill the polygon
        if bbox_on:
            #--- bbox 
            bbox = anno['bbox']
            #print(bbox)
            x, y, width, height = bbox
            ax = plt.gca()
            rect = patches.Rectangle((x, y), width, height, linewidth=2, edgecolor='blue', facecolor='lightblue', alpha=0.9)
            ax.add_patch(rect)

plt.imshow(img)
plt.show()
```
